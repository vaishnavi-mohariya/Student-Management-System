```python
class Students_information:
    def __init__(self):
        self.students_details={}
    def creation(self):
        while True:
            userinput=input('Do you want to enter details of student: Yes/No')
            
            if(userinput=='no'):
                break
            elif(userinput=='yes'):
                name=input('Enter the Name of Student:')
                rollnumber=int(input("Enter Student Roll Number:"))
                marks=int(input("Enter the Marks of Student:"))
                
                self.students_details[rollnumber]={'Name':name,'Marks':marks}
                
            else:
                print("invalid choice")
    def showdetails(self):
        students_record=[]
        for roll,details in self.students_details.items():
            students_info={
                'Rollnumber':roll,
                'Name':details['Name'],
                'Marks':details['Marks']
            }
            students_record.append(students_info)
        return students_record
        
    def searchdetails(self,rollnumber):
        if rollnumber in self.students_details:
            return self.students_details[rollnumber]
        else:
            print("student not found")
    def updatemarks(self):
        rollnumber=int(input("Enter the Rollnumber to update marks"))
        if rollnumber in self.students_details:
            newmarks=int(input("Enter new marks to update"))
            self.students_details[rollnumber]['Marks'] = newmarks
            print("Marks updated succcesfully")
            print(self.students_details)
        else:
            print("Student not found")
            
class Mainmenu:
    def __init__(self):
        self.studentmanager=Students_information()
    def show_menu(self):
        while True:
            print("\n Students Management System")
            print("---------------------------------")
            
            print("1: Add students details")
            print("2: Show All Students")
            print("3: Search Student by Roll Number")
            print("4: Update Student Marks")
            print("5: Exit")
            choice=int(input("Select an option from above."))
            if choice==1:
                self.studentmanager.creation()
            elif choice==2:
                students=self.studentmanager.showdetails()
                print("------------------------------------------------------------------------------------------------")
                print(students)
                print("-------------------------------------------------------------------------------------------------")
            elif choice==3:
                rollnum=int(input("enter roll number"))
                print(self.studentmanager.searchdetails(rollnum))
            elif choice==4:
                self.studentmanager.updatemarks()
            elif choice==5:
                print("Thank you! Exiting the program....")
                break
            else:
                print("Invalid choice")
menu=Mainmenu()
menu.show_menu()
```
